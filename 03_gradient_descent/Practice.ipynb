```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#%matplotlib inline
import math
def gradient_descent(x,y):
    m_curr= b_curr=0
    learning_rate=0.0002
    n = len(x)
    iterations=1000000
    cost=0
    last_i=0
    #plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(iterations):
        #y=mx+b
        y_predicted = m_curr * x + b_curr
        current_cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        
        #partial derivative
        md  = -(2/n) * sum(x*(y-y_predicted))
        mb  = -(2/n) * sum(y-y_predicted)
        #updated slope & interceptor
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * mb
        last_i+=1;
        if(math.isclose(cost,current_cost,rel_tol=1e-20)) :
            break;
        cost=current_cost

    print("slope(m): {},\t intercept(b): {}, \t cost {},\t\t iteration {}".format(m_curr,b_curr, cost,last_i))

```

```python
#For Testing, consider
x1= np.array([1,2,3,4,5])
y1 = np.array([5,7,9,11,13])
```

```python
gradient_descent(x1,y1)
```
